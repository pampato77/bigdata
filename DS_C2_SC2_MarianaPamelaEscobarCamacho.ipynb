```python
import numpy as np
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

vacunas = pd.read_csv('/content/drive/MyDrive/BigDataEjercicios/country_vaccinations.csv')
```

b. Mostrar la estructura y tipos de datos de cada columna para identificar que operaciones puedes realizar con cada una de ellas, asegurandote que la columna con fechas sean del tipo datetime64

```python
vacunas ["date"] = vacunas["date"].astype("datetime64")
vacunas.dtypes
```

c. Determinar la cantidad de vacunas aplicadas de cada compañia, como lo reporta cada país en la columna vaccines, agrupar por vacunas y realizar la sumatoria.

```python
vacunas1 = vacunas[['vaccines','people_vaccinated']].groupby('vaccines').sum()
vacunas1
```

d. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
vacunas2 = vacunas['people_vaccinated'].sum()
vacunas2
```

e. Calcular el promedio de vacunas aplicadas por pais

```python
vacunas3 = vacunas[['country','daily_vaccinations']].groupby('country').mean()
vacunas3
```

f. Determinar al cantidad de vacunas aplicadas el dìa 29/01/21 en todo el mundo

```python
vacunas4 = vacunas.loc[vacunas['date'] == '2021-01-29']['total_vaccinations'].sum()
vacunas4
```

g. Crear un dataframe nuevo denominado ConDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw

```python
conDiferencias = vacunas.assign( diferencias = vacunas['daily_vaccinations'] - vacunas['daily_vaccinations_raw'])
conDiferencias
```

h. Obtener el periodo de tiempo entre el registro con fechas más recientes y el registro con fecha mas antigua.

```python
periodo = vacunas['date'].max() - vacunas['date'].min()
periodo
```

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada dia (usar la columna vaccines y separar por el carácter,).

```python
conCantidad = vacunas.assign(canVac= vacunas['vaccines'].str.split('/ ').str.len())
conCantidad
```

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = vacunas[vacunas["date"]<'2020-12-20']
antes20
```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer

```python
pfizer = vacunas[vacunas["vaccines"].str.contains('Pfizer')]
pfizer
```

l. Almacenar los dataframes generados ( conDiferencias, conCantidad, antes20 y pfizer) en un archivo de excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.

```python
resultados = pd.ExcelWriter('resultadosReto.xlsx')
conDiferencias.to_excel(resultados, sheet_name='diferencias', index=False)
conCantidad.to_excel(resultados, sheet_name='cantidad', index=False)
antes20.to_excel(resultados, sheet_name='antes20', index=False)
pfizer.to_excel(resultados, sheet_name='pfizer', index=False)
resultados.save()
```
